# Interview Transcription & Speaker Diarization

Transcribe audio interviews with speaker identification:

1. **WhisperX** — transcription + forced alignment + speaker diarization
2. **LLM post-processing** — punctuation, filler word removal, ASR error correction
3. **Output** — speaker-labeled transcript (.txt), subtitles (.srt), raw segments (.json)

This notebook is a thin launcher — all logic lives in the `interview_transcription` package.


## Prerequisites

1. **Hugging Face token** — required for pyannote speaker diarization models
   - Create a token at [hf.co/settings/tokens](https://huggingface.co/settings/tokens)
   - Accept the terms for these models:
     - [pyannote/segmentation-3.0](https://huggingface.co/pyannote/segmentation-3.0)
     - [pyannote/speaker-diarization-3.1](https://huggingface.co/pyannote/speaker-diarization-3.1)
2. **LLM API key** (optional) — for transcript post-processing
   - Anthropic API key ([console.anthropic.com](https://console.anthropic.com)) **or**
   - OpenAI API key ([platform.openai.com](https://platform.openai.com))
3. **Audio file** on Google Drive

Add secrets (`HF_TOKEN`, `ANTHROPIC_API_KEY` or `OPENAI_API_KEY`) via Colab's Secrets panel (key icon in sidebar).
